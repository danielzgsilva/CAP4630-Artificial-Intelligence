```
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

## Average MNIST Digits

```
from keras.datasets import mnist

(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```
x = np.concatenate((x_train, x_test))
y = np.concatenate((y_train, y_test))
```

```
print(x.shape)
print(y.shape)
```

```
# Creating number dictionary
nums = {i: [] for i in range(10)}
```

```
# Placing images of each digit into their respective category in the dictionary
for i in range(len(y)):
  nums[y[i]].append(x[i])
```

```
# Just casting images into numpy arrays
for i in range(len(nums)):
  nums[i] = np.array(nums[i])
```

```
# Average each category in the dictionary
avg_nums = {i: np.mean(nums[i], axis=0) / 255 for i in range(len(nums))}
```

```
# Plotting average numbers
fig = plt.figure(10, (14,14))

for i in range(len(avg_nums)):
    ax = plt.subplot(2,5,i+1)
    ax.imshow(avg_nums[i], cmap='Greys')
    ax.set_title('Average {}'.format(i))
```
